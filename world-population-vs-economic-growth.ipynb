```python
import numpy as np
import pandas as pd
WP=pd.read_csv("/kaggle/input/world-population-dataset/world_population.csv")
WP
```

```python
import warnings
warnings.simplefilter("ignore")
import geopandas as gpd
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
WP['Country/Territory'] = WP['Country/Territory'].replace({'United States': 'United States of America'})
WP['Country/Territory'] = WP['Country/Territory'].replace({'DR Congo': 'Democratic Republic of Congo'})
```

```python
WE= pd.read_csv("/kaggle/input/global-economy-indicators/Global Economy Indicators.csv")
WE
```

# Null Values and filling them

```python
for col in WP.columns: print(f'{col}: {WP[col].isnull().mean() * 100:.2f}% null')
```

```python
for col in WE.columns: print(f'{col}: {WE[col].isnull().mean() * 100:.2f}% null')
```

```python
for col in WE.columns: WE[col] = WE[col].fillna("NA" if WE[col].dtype == 'O' else WE[col].mean())
```

# World Population Growth, Density, Rate and Percentage

```python
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
world = pd.merge(world, WP[['Country/Territory', 'Growth Rate']], 
                 how='left', left_on='name', right_on='Country/Territory')

fig = px.choropleth(world,
                    locations='iso_a3',
                    color='Growth Rate',
                    hover_name='name',
                    color_continuous_scale='OrRd',
                    title='World Population Growth Rate by Country')

fig.show()
```

```python
world = pd.merge(world, WP[['Country/Territory', 'Density (per km²)']], 
                 how='left', left_on='name', right_on='Country/Territory')

fig = px.choropleth(
    world,
    locations='iso_a3',
    color='Density (per km²)',
    hover_name='name',
    color_continuous_scale='Viridis',
    title='Population Density Across Continents'
)

fig.update_geos(
    projection_type='natural earth',
    showcoastlines=True,
    coastlinecolor='Black',
    showland=True,
    landcolor='white'
)

fig.show()
```

```python
top_countries = WP.sort_values(by='2022 Population', ascending=False).head(10)
fig_map_top = px.choropleth(
    top_countries,
    locations='CCA3',
    color='2022 Population',
    hover_name='Country/Territory',
    title='Top 10 Countries by Population',
    color_continuous_scale='Viridis',
    projection='natural earth',
)
fig_map_top.show()

bottom_countries = WP.sort_values(by='2022 Population').head(10)

fig_bottom = px.bar(
    bottom_countries,
    x='Country/Territory',
    y='2022 Population',
    title='Bottom 10 Countries by Population',
    labels={'2022 Population': 'Population'},
)
fig_bottom.show()
```

```python
fig_world_population = px.choropleth(
    WP,
    locations="CCA3",
    color="World Population Percentage",
    hover_name="Country/Territory",
    title="World Population Percentage by Country",
    color_continuous_scale=px.colors.sequential.Plasma,
    projection="natural earth",
)

fig_world_population.show()
```

```python
fig = px.sunburst(
    WP,
    path=['Continent', 'Country/Territory'],
    values='2022 Population',
    width=1000, 
    height=800,
    title='Population Distribution (2022)'
)

fig.show()
```

```python
fig_line_population = px.line(
    WE,
    x=' Year ',
    y=' Population ',
    color=' Country ',
    title='Population Growth Over Time',
)

fig_line_population.show()
```

# GDP

```python
fig_gdp_growth = px.line(
    WE,
    x=' Year ',
    y=' Gross Domestic Product (GDP) ',
    color=' Country ',
    title='GDP Growth Over Time',
)
fig_gdp_growth.show()


top_countries = WE.groupby(' Country ')[' Gross Domestic Product (GDP) '].max().sort_values(ascending=False).head(10)
top_countries_list = top_countries.index.tolist()
top_countries_df = WE[WE[' Country '].isin(top_countries_list)]
fig_gdp_growth10 = px.bar(
    top_countries_df,
    x=' Year ',
    y=' Gross Domestic Product (GDP) ',
    color=' Country ',
    title='GDP Growth Over Time (Top 10 Countries)',
)
fig_gdp_growth10.show()
```

# Interactive Bar Graph for GDP and Population Growth

## Use the below code to change in GDP Growth Rate and Population Growth rate, then next cell can be used to visualize it

```python
WE['GDP Growth'] = WE.groupby(' Country ')[' Gross Domestic Product (GDP) '].pct_change() * 100
WE['Population Growth'] = WE.groupby(' Country ')[' Population '].pct_change() * 100

growing_countries = WE[(WE['GDP Growth'] > 5) & (WE['Population Growth'] > 5)] #Change in the percentage here, for both GDP and Population

top_growing_countries = growing_countries.groupby(' Country ')[['GDP Growth', 'Population Growth']].mean().sort_values(by=['GDP Growth', 'Population Growth'], ascending=False).head(10)
top_growing_countries
```

```python
fig = px.bar(
    top_growing_countries.reset_index(),
    x=' Country ',
    y=['GDP Growth', 'Population Growth'],
    title='Top Growing Countries - GDP and Population Growth',
    labels={'value': 'Percentage Growth'},
)

fig.show()
```
